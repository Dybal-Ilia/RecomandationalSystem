```python
from sklearn.model_selection import train_test_split
users = df.UserId.unique()
products = df.ProductId.unique()

userid2idx = {o:i for i,o in enumerate(users)}
product2idx = {o:i for i,o in enumerate(products)}
df['UserId'] = df['UserId'].apply(lambda x: userid2idx[x])
df['ProductId'] = df['ProductId'].apply(lambda x: product2idx[x])

train, test = train_test_split(df[['UserId', 'ProductId', 'Score']], test_size = 0.2, shuffle = True)
```

```python
from keras.layers import Input, Embedding, Flatten, dot, Dense, Dropout
from keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
n_products = len(df['ProductId'].unique())
n_users = len(df['UserId'].unique())
n_latent_factors = 24
user_input = Input(shape = (1, ), name = 'user_input', dtype = 'int64')
user_embedding = Embedding(n_users, n_latent_factors, name = 'user_embedding')(user_input)
user_vec = Flatten(name = 'FlattenUsers')(user_embedding)

product_input = Input(shape = (1, ), name = 'product_input', dtype = 'int64')
product_embedding = Embedding(n_products, n_latent_factors, name = 'product_embedding')(product_input)
product_vec = Flatten(name = 'FlattenProducts')(product_embedding)

sim = dot([user_vec, product_vec], name = 'Similarity-Dot-Product', axes = 1)
model = Model([user_input, product_input], sim)
model.compile(loss = 'mse')
early_stopping = EarlyStopping(
    monitor = 'val_loss',
    patience = 3,
    restore_best_weights = True
)
```

```python
history = model.fit([train.UserId, train.ProductId], train.Score, batch_size = 256, epochs = 25,
                    validation_data = ([test.UserId, test.ProductId], test.Score), callbacks = [early_stopping], verbose = 1)
```

```python
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 10, 5
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], 'g')
plt.plot(history.history['val_loss'], 'b')
plt.title('График ошибки')
plt.ylabel('Ошибка')
plt.xlabel('Эпоха')
plt.legend(['Ошибка на обучении', 'Ошибка на валидации'], loc = 'upper right')
plt.grid(True)
plt.show()
```

```python
from keras import regularizers
user_input = Input(shape = (1, ), name = 'user_input', dtype = 'int64')
user_embedding = Embedding(n_users, n_latent_factors, name = 'user_embedding', embeddings_regularizer = 'l2')(user_input)
user_vec = Flatten(name = 'FlattenUsers')(user_embedding)
product_input = Input(shape = (1, ), name = 'product_input', dtype = 'int64')
product_embedding = Embedding(n_products, n_latent_factors, name = 'product_embedding', embeddings_regularizer = 'l2')(product_input)
product_vec = Flatten(name = 'FlattenProducts')(product_embedding)
sim = dot([user_vec, product_vec], name = 'Similarity_Dot_Product', axes = 1)
nn_inp = Dense(8, activation = 'relu', kernel_regularizer = regularizers.L2(0.01), activity_regularizer = regularizers.L2(0.01))(sim)
nn_inp = Dropout(0.25)(nn_inp)
nn_inp = Dense(16, activation = 'relu', kernel_regularizer = regularizers.L2(0.01), activity_regularizer = regularizers.L2(0.01))(nn_inp)
nn_inp = Dropout(0.25)(nn_inp)
nn_inp = Dense(1, activation = 'relu')(nn_inp)
nn_model = Model([user_input, product_input], nn_inp)
nn_model.compile(loss = 'mse')
nn_model.summary()
```

```python
history = nn_model.fit([train.UserId, train.ProductId], train.Score, batch_size = 256, epochs = 25,
                    validation_data = ([test.UserId, test.ProductId], test.Score), callbacks = early_stopping, verbose = 1)
```

```python
plt.plot(history.history['loss'], 'g')
plt.plot(history.history['val_loss'], 'b')
plt.title('График ошибки')
plt.ylabel('Ошибка')
plt.xlabel('Эпоха')
plt.legend(['Ошибка на обучении', 'Ошибка на валидации'], loc = 'upper right')
plt.grid(True)
plt.show()
```
